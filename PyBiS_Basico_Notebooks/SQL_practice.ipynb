# Generamos la data

```python
import random 
import pandas as pd
import faker
import sqlite3 as sql

fk = faker.Faker()
```

```python
n = 1000

regiones = ['Norte', 'Sur', 'Este', 'Oeste']
ID_jefe = [random.randint(1, 4) for _ in range(n)]

# Usamos la libreria para generar nombres y puestos aleatorios
nombres = [fk.name() for _ in range(n)]
edades = [random.randint(18, 65) for _ in range(n)]
salarios = [random.uniform(1000, 10000) for _ in range(n)]
puestos = [fk.job() for _ in range(n)]
ciudades = [fk.city() for _ in range(n)]
region = [random.choice(regiones) for _ in range(n)]
ID_jefes = [random.choice(ID_jefe) for _ in range(n)]

# creamos el dataframe
df = pd.DataFrame({
    'nombre': nombres,
    'edad': edades,
    'salario': salarios,
    'puesto': puestos,
    'ciudad': ciudades,
    'region': region,
    'ID_jefe': ID_jefes
})

# Creamos la base de datos
conn = sql.connect('Practica_sql.db')

# Realizamos las inserciones
df.to_sql('datos', conn, if_exists='append', index=False)

# Creamos la informacion de la tabla jefes
nombres_jefes = ['Juan', 'Maria', 'Pedro', 'Ana']
edades_jefes = [random.randint(40, 60) for _ in range(4)]

df_jefes = pd.DataFrame({
    'ID_jefe': [1, 2, 3, 4],
    'nombre': nombres_jefes,
    'edad': edades_jefes
})

df_jefes.to_sql('jefes', conn, if_exists='append', index=False)
```

```python
# Realizamos la consulta a la tabla de datos
query = """
SELECT * FROM datos LIMIT 5;
"""
con = sql.connect('Practica_sql.db')
query_df = pd.read_sql_query(query, con)
con.close()

query_df
```

```python
# Realizamos la consulta a la tabla de jefes

query_df2 = """
SELECT * FROM jefes;
"""

con = sql.connect('Practica_sql.db')
query_df2 = pd.read_sql_query(query_df2, con)
con.close()

query_df2
```

# Ejercicios

1. Obtén la suma de los salarios para aquellos empleados de la región Este.

```python
query_sol1 = """
SELECT 
    region,
    SUM(salario) as suma_salarios 
FROM 
    datos 
WHERE   
    region = 'Este' 
GROUP BY 
    region;
"""

con = sql.connect('Practica.db')
query_df = pd.read_sql_query(query_sol1, con)
con.close()

query_df
```

2. Obtener el segundo màs alto de tal cosa

```python
query_sol1 = """
SELECT 
    nombre, Max(salario) as salario
FROM 
    datos
WHERE
    nombre != (SELECT nombre FROM datos ORDER BY salario DESC LIMIT 1)
"""

con = sql.connect('Practica.db')
query_df = pd.read_sql_query(query_sol1, con)
con.close()

query_df
```
